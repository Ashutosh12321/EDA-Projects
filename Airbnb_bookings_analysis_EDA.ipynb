## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1

```python
airbnb_df.head()
```

```python
# Chart - 1 visualization code
# Bar chart - Average price of listings vs neighbourhood_group

# Calculate the average price of each neighbourhood group
neighbourhood_group = airbnb_df.groupby("neighbourhood_group").agg({"price" : "mean"}).reset_index()

# Sort the data in descending order to show the most expensive neighborhoods first
neighbourhood_group.sort_values("price", ascending = False, inplace = True)

# Setting chart size
plt.figure(figsize = (10, 5))

# Setting axis labels and title
plt.title("Average price of listings vs neighbourhood_group")
plt.xlabel("Neighborhood_group")
plt.ylabel("Average Price")

# Creating the plot
sns.barplot(data = neighbourhood_group, x = "neighbourhood_group", y = "price", width = 0.20, bottom = 0, label = ["Manhattan", "Brooklyn", "Staten Island", "Queens", "Bronx"])
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility

# Show the legend
plt.legend()

# Show the bar chart
plt.show()
```

##### 1. Why did you pick the specific chart?

* We need to show the average price for different categories here categories are neighbourhood_groups.
* Bar chart is used to display categorical data with different rectangular bars.

##### 2. What is/are the insight(s) found from the chart?

* Maximum average price is for Manhattan neighbourhood group and minimum average price is for Bronx neighbourhood group.
* According to the average price the most expensive neighbourhood group is Manhattan and the most affordable neighbourhood group is Bronx.

#### Chart - 2

```python
# Chart - 2 visualization code
# Pie chart - Distribution of room types

# Using a plot style
plt.style.use("seaborn")

# Type count
type_count = [1, 2, 3]

# Extracting the types of rooms from the dataset
room_types = airbnb_df["room_type"]


# Setting radius and explode for better visualization
radius = 0.8
explode = [0, 0.01, 0.01]

# Setting axis labels and chart title
labels = ["Entire home/apt", "Private room", "Shared room"]
plt.title("Distribution of room types")

# Creating plot
plt.pie(type_count, data = room_types, radius = radius, explode = explode, labels = labels, autopct = "%.2f%%")

# Show rhe legend
plt.legend()

# Show the pie chart
plt.show()
```

##### 1. Why did you pick the specific chart?

* Pie chart is used here to display the percentage of different types of rooms available.

##### 2. What is/are the insight(s) found from the chart?
